### DGP

```python
tau = 2 # Traetment Effect
corr = .5 # Correlation in between the covariates
p = 10 # number of covariates generated
p0 = 0 # number of covariates used in the DGP
Nrange = range(10,1000,2) # loop over N values

(nvalues,tauhats,sehats,lb,ub) = mf.fn_run_experiments(tau,Nrange,p,p0,corr,conf=False,flagX=False)
```

```python
# plot the estimated value for the treatment effect, actual treatment effect value and confidence interval
caption = """Estimates of the treatment effect parameter 
    for a randomized experiment without confounders"""

mf.fn_plot_with_ci(nvalues,tauhats,tau,lb,ub,caption)
```

### Monte Carlo Experiment 
### For sample N=100 and N=1000

```python
estDict = {}
R = 2000
for N in [100,1000]:
    tauhats = []
    sehats = []
    for r in tqdm(range(R)):
        Yexp,T = mf.fn_generate_data(tau,N,p,p0,corr,conf=False,flagX=False)
        Yt = Yexp[np.where(T==1)[0],:] # outcome of the treated
        Yc = Yexp[np.where(T==0)[0],:] # outcome of the controlled
        tauhat,se_tauhat = mf.fn_tauhat_means(Yt,Yc) # estimate tau and SE values for each experiment
        tauhats = tauhats + [tauhat] # store estimated value of treatment effects
        sehats = sehats + [se_tauhat] # store estimated value of SE 
    estDict[N] = {
        'tauhat':np.array(tauhats).reshape([len(tauhats),1]),
        'sehat':np.array(sehats).reshape([len(sehats),1])
    }

# caculate bias, RMSE and size for N=100 and N=1000
tau0 = tau*np.ones([R,1])
for N, results in estDict.items():
    (bias,rmse,size) = mf.fn_bias_rmse_size(tau0,results['tauhat'],
                                         results['sehat'])
    print(f'N={N}: bias={bias}, RMSE={rmse}, size={size}')
```

## Exporting Generated Data

```python
data =np.concatenate([Yexp,T],axis = 1)
Data_1a = pd.DataFrame(data,columns = ['Outcome (Y)','Treatment (T)'])
Data_1a.to_csv("Simulated_Data_1a")# Export to CSV
Data_1a
```

### DGP

```python
tau = 2 # Traetment Effect
corr = .5 # Correlation in between the covariates
p = 10 # number of covariates generated
p0 = 5 # number of covariates used in the DGP
Nrange = range(10,1000,2) # loop over N values

(nvalues2,tauhats2,sehats2,lb2,ub2) = mf.fn_run_experiments(tau,Nrange,p,p0,corr,conf=False,flagX=1)
```

```python
caption = """Estimates of the treatment effect parameter 
    for a randomized experiment without confounders"""

mf.fn_plot_with_ci(nvalues2,tauhats2,tau,lb2,ub2,caption)
```

### Monte Carlo Experiment 
### For sample N=100 and N=1000

```python
estDict = {}
R = 2000
for N in [100,1000]:
    tauhats = []
    sehats = []
    for r in tqdm(range(R)):
        Yexp,T,X = mf.fn_generate_data(tau,N,p,p0,corr,conf=False,flagX=1)
        Xobs = X[:,:p0]
        covars = np.concatenate([T,Xobs],axis = 1) #concatenate covariates
        mod = sm.OLS(Yexp,covars) # run OLS regression
        res = mod.fit() # residuals
        tauhat = res.params[0] # estimated treatment effect
        se_tauhat = res.HC1_se[0] # estimated SE
        tauhats = tauhats + [tauhat] # store estimated value of treatment effects
        sehats = sehats + [se_tauhat] # store estimated value of SE 
    estDict[N] = {
        'tauhat':np.array(tauhats).reshape([len(tauhats),1]),
        'sehat':np.array(sehats).reshape([len(sehats),1])
    }

# caculate bias, RMSE and size for N=100 and N=1000
tau0 = tau*np.ones([R,1])
for N, results in estDict.items():
    (bias,rmse,size) = mf.fn_bias_rmse_size(tau0,results['tauhat'],
                                         results['sehat'])
    print(f'N={N}: bias={bias}, RMSE={rmse}, size={size}')   
```

## Exporting Generated Data

## Real-Life Example

State of California introduced an incentive program called "Clean Vehicle Rebate Program (CVRP)" that aims to reduce Co2 emissions by offering a discount incentive (rebates) to increase EV sales.

The effectiveness of the program is measured by how much EV cars sales changed (increased or decreases) every year compared to the previous year. If only the treatment (CVRP) included in the analysis, the obtained Treatment effect will be biased as there are other factors that affect the sales of EVs. For example, gas prices is an essential part of the analysis and has a direct causal effect on EV sales.  

In 2020 and 2021, gas prices fall to a historical level due to Covid-19 pandemic and that certainly slow down consumers desire for EVs while in 2022 gas prices raised dramatically which certainly will motivate people to purchase electric vehicles.
So, if gas prices has not been included in the analysis (as in setting 1 part a) that will underestimate the progam effect (Treatment) in 2020 and 2021 while overestimate its effect in 2022.

```python
g = gr.Digraph()
g.edge("CVRP (T)", "EV Purchases (Y)")
g.edge("Gas Prices (X)", "EV Purchases (Y)")
g.node("Gas Prices (X)", "Gas Prices (X)", color="red")
g
```

## Real-Life Example

Most of the companies use targeted advertising  campaigns to boost sales of products or services. As a start a pilot randomized experiment will be launched to measure the effectiveness of the ad on revenue. However, if the ad never seen by the consumer or delivered to the wrong audience that will lead to underestimating the treatment effect the ad (as in setting 3 part a) while the issue is not the content of the ad but the delivery method used to reached the consumer. So, all of the methods of delivery should be included in the analysis to get an unbiased. estimates of the advertisement campaign effectiveness on revenue.
